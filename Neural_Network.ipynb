```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
y=np.array(df['label'])
X=df[['air_pollution',
       'dust_allergy', 'occupational_hazards', 'chronic_lung_disease',
       'balanced_diet', 'smoking',
       'passive_smoker', 'chest_pain', 'fatigue',
       'weight_loss', 'shortness_of_breath','wheezing',
       'swallowing_difficulty', 'clubbing_of_finger_nails', 'frequent_cold',
       'dry_cough', 'snoring']]
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.layers import Dropout
from keras import regularizers
from tensorflow.keras.layers import Activation
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, df['label'], test_size=0.3, random_state=42)
```

```python
y_train
```

```python
X_train
```

```python
y_test
```

```python
X_test
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python

# Build the model
model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary(0,'')
# Train the model
history = model.fit(X_train, y_train, epochs=5, batch_size=100, validation_data=(X_test, y_test))
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}')
```

```python
plt.plot(history.history['loss'], label='training data')
plt.plot(history.history['val_loss'], label='validation data')
plt.legend()
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.show()
```

```python
plt.plot(history.history['accuracy'], label='Accuracy training data')
plt.plot(history.history['val_accuracy'], label='Accuracy validation data')
plt.legend()
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
threshold = 0.63

y_pred = model.predict(X_test)
y_pred_binary = (y_pred > threshold).astype(int)
plt.figure(figsize=(8, 6))
CM = confusion_matrix(y_test, y_pred_binary)
display = ConfusionMatrixDisplay(confusion_matrix=CM)
display.plot()
plt.title("Confusion Matrix")
plt.show()
```
