```python
os.listdir(data_dir)
```

```python
data = pd.read_csv('/content/Cleaned cancer patient data sets.csv')
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
#Dropping unneccessary columns
data.drop(["index", "Patient Id"], axis=1, inplace = True)
```

```python
# Cleaning Column Names
data.rename(columns=str.lower, inplace=True)
data.rename(columns={col: col.replace(" ", "_") for col in data.columns}, inplace=True)
```

```python
data.info()
```

```python
cancer_patients = pd.DataFrame(data)
cancer_patients.to_csv("Cleaned cancer patient data sets.csv", index= False)
```

```python
#correlation Heat Map
plt.subplots(figsize = (24, 24))
sns.heatmap(data.corr(), annot = True)
```

```python
data['level'].value_counts()
```

```python
sea = sns.FacetGrid(data, col = "level", height = 4)
sea.map(sns.distplot, "age")
```

```python
sea = sns.FacetGrid(data, col = "level", height = 4)
sea.map(sns.distplot, "gender")
```

```python
# Lung cancer chances due to Air pollution
plt.figure(figsize = (11, 9))
plt.title("Lung Cancer Chances Due to Air Polution")
plt.pie(data['level'].value_counts(), explode = (0.1, 0.02, 0.02), labels = ['High', 'Medium', 'Low'], autopct = "%1.2f%%", shadow = True)
plt.legend(title = "Lung Cancer Chances", loc = "lower left")

```
